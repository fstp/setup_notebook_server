```python
from numpy.polynomial import Polynomial
```

```python
p = Polynomial([1, -2, 3])
x, y = p.linspace(n=100, domain=[5, -5])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title(f'P(x) = {p}')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y, x)
dx = x[1] - x[0]
y_int = np.cumsum(y) * dx
```

```python
#1
theta = np.linspace(0, 2*np.pi, 1_000)
r = 1 + 3/4*np.sin(3*theta)
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
plt.plot(x, y);
```

```python
#2
A = 1/2 * sum(r**2) * (theta[1] - theta[0])
print(A)
```

```python
#3
integrand = np.sqrt(r**2 + np.gradient(r, theta)**2)
integrand[:10]
```

```python
dtheta = theta[1] - theta[0]
print(sum(integrand) * dtheta)
```
